```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
```

```python
# a)   
fig11 = plt.figure(num=11, figsize=(12, 8))
ax11 = fig11.add_subplot(1, 1, 1)
ax11.set_xlim(left=-8, right=8)
ax11.set_ylim(bottom=-2, top=6)
ax11.set_aspect('equal')
ax11.grid(True, alpha=0.3)

# 1. 
circle = mpatches.Circle(xy=(-6, 4), radius=1.5, color='blue', 
                        linestyle='-', linewidth=2, fill=False)

# 2. 
ellipse = mpatches.Ellipse(xy=(-3, 4), width=3.0, height=1.5, 
                          color='green', linestyle='--', fill=False)

# 3. 
rectangle = mpatches.Rectangle(xy=(0, 2.5), width=2, height=1.5, 
                              angle=60, fill=False, color='purple')

# 4.  
arc = mpatches.Arc(xy=(3, 4), width=4, height=2, angle=30,
                  theta1=45, theta2=270, color='orange', linewidth=2)

# 5.  
wedge = mpatches.Wedge(center=(6, 4), r=1.5, theta1=90, theta2=180, 
                      fill=False, color='brown')

# 6. 
trapezoid_points = np.array([[-7, -1], [-5, -1], [-3, 1], [-9, 1]])
polygon = mpatches.Polygon(xy=trapezoid_points, closed=True, 
                          fill=False, color='red', linewidth=2)

# 7.  ,  
arrow = mpatches.Arrow(x=-2, y=-1, dx=3, dy=0, color='red', width=0.2)

# 8.      
fbox1 = mpatches.FancyBboxPatch(xy=(1, -1), width=4, height=3, 
                               boxstyle='round,pad=0.1', fill=False)
fbox2 = mpatches.FancyBboxPatch(xy=(1.5, -0.5), width=3, height=2, 
                               boxstyle='round,pad=0.1', fill=False)
fbox3 = mpatches.FancyBboxPatch(xy=(2, 0), width=2, height=1, 
                               boxstyle='round,pad=0.1', fill=False)

#     
primitives = [circle, ellipse, rectangle, arc, wedge, polygon, arrow, 
             fbox1, fbox2, fbox3]

for primitive in primitives:
    ax11.add_patch(primitive)

#  
ax11.text(-6, 6, '', ha='center')
ax11.text(-3, 6, '', ha='center')
ax11.text(0, 4.5, '', ha='center')
ax11.text(3, 6, '', ha='center')
ax11.text(6, 6, '', ha='center')
ax11.text(-6, -1.5, '', ha='center')
ax11.text(-0.5, -1.5, '', ha='center')
ax11.text(3, -1.5, '\n', ha='center')

plt.title('  Matplotlib')
plt.tight_layout()
plt.show()
```

```python
# b)      Line2D
fig12 = plt.figure(num=12, figsize=(10, 4))
ax12 = fig12.add_subplot(1, 1, 1)
ax12.set_xlim(left=0, right=4*np.pi)
ax12.set_ylim(bottom=-1.1, top=1.1)

#    
x_data = np.linspace(0, 4*np.pi, 100)
y_data = np.sin(x_data)

#     Line2D
sin_line = mlines.Line2D(xdata=x_data, ydata=y_data, 
                        linewidth=2, color='blue', 
                        marker='o', markersize=3, markevery=5)

ax12.add_line(sin_line)
ax12.set_xlabel('x')
ax12.set_ylabel('sin(x)')
ax12.set_title(' ')
ax12.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# c)      
fig13 = plt.figure(num=13, figsize=(10, 8))
ax13 = fig13.add_subplot(1, 1, 1)
ax13.set_xlim(0, 1.5)
ax13.set_ylim(0, 1.5)

points = [(0.25, 0.25), (0.50, 0.50), (0.75, 0.75), (1.0, 1.0), (1.25, 1.25)]
xs = [p for (p, _) in points]
ys = [p for (_, p) in points]

ax13.plot(xs, ys, marker='o', markersize=8, linestyle='None', color='blue')

#   
x, y = points[0]
ax13.annotate(' 1', xy=(x, y), xytext=(20, 20),
             textcoords='offset points', 
             bbox=dict(boxstyle='round', fc='lightblue'),
             arrowprops=dict(arrowstyle='->', color='gray'))

x, y = points[1]
ax13.annotate(' 2', xy=(x, y), xytext=(0, 30),
             textcoords='offset points', rotation=90,
             bbox=dict(boxstyle='round', fc='lightgreen'))

x, y = points[2]
ax13.annotate(' 3', xy=(x, y), xytext=(-40, 30),
             textcoords='offset points',
             bbox=dict(boxstyle='round', fc='yellow', alpha=0.7),
             fontstyle='italic', fontsize=12)

x, y = points[3]
ax13.annotate(' 4', xy=(x, y), xytext=(30, -40),
             textcoords='offset points',
             bbox=dict(boxstyle='round', fc='orange'),
             fontweight='bold')

x, y = points[4]
ax13.annotate(r'$\mathscr{P}(1.25, 1.25)$', xy=(x, y), 
             xytext=(-60, -60), textcoords='offset points',
             bbox=dict(boxstyle='round', fc='pink'))

#  
ax13.annotate(r'$\lim_{x \to +\infty}\left(\frac{\pi(x)}{x/\ln x}\right)=1$', 
             xy=(0.3, 1.3), xytext=(0, 0), textcoords='offset points',
             fontsize=16, bbox=dict(boxstyle='round', fc='white'))

plt.title('   ')
plt.tight_layout()
plt.show()
```

```python
# d)    
fig15 = plt.figure(num=15, figsize=(8, 8))
ax15 = fig15.add_subplot(1, 1, 1)

#   : x = R*cos(t), y = R*sin(t)
t = np.linspace(0, 2*np.pi, 50)
R = 3.0
x = R * np.cos(t)
y = R * np.sin(t)

#  ( )
dx = -R * np.sin(t)  #   x
dy = R * np.cos(t)   #   y

#     
length = np.sqrt(dx**2 + dy**2)
dx_norm = 0.5 * dx / length
dy_norm = 0.5 * dy / length

#    ( 4-  )
ax15.quiver(x[::4], y[::4], dx_norm[::4], dy_norm[::4], 
           units='xy', angles='xy', scale=1.0, scale_units='xy',
           width=0.03, color='red', label=' ')

#   
ax15.plot(x, y, linewidth=2, color='blue', label='')

ax15.set_xlabel(r' $Ox$')
ax15.set_ylabel(r' $Oy$')
ax15.set_title('   ')
ax15.legend()
ax15.set_aspect('equal')
ax15.grid(True, alpha=0.3)

#   
ax15.plot(0, 0, 'ko', markersize=5)
ax15.annotate('', xy=(0, 0), xytext=(0.5, 0.5),
             arrowprops=dict(arrowstyle='->'))

plt.tight_layout()
plt.show()
```
