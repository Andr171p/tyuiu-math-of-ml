```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  
t = np.linspace(-4*np.pi, 4*np.pi, 400)
x = np.sin(t)
y = np.cos(t)

#     
fig01 = plt.figure(num=0, figsize=(10, 6))
ax01 = fig01.add_subplot(2, 1, 1)
ax02 = fig01.add_subplot(2, 1, 2)

#      
ax01.plot(t, x, linewidth=2, linestyle="--", color="red")
ax02.plot(t, y, linewidth=1.5, linestyle="-.", color="blue")

#    
ax01.set_title(" ")
ax02.set_title(" ")
ax01.set_ylabel("sin(t)")
ax02.set_ylabel("cos(t)")
ax02.set_xlabel("t")

plt.tight_layout()
plt.show()
```

```python
t = np.linspace(-4*np.pi, 4*np.pi, 400)
x = np.sin(t) + np.cos(t) + np.cos(2*t)
y = np.cos(2*t) + np.sin(np.cos(t))

fig04 = plt.figure(10)
ax04 = fig04.add_subplot(1, 1, 1)

label_01 = r'$\sin({t}) + \cos({t}) + \cos({2t})$'
label_02 = r'$\cos{(2t)} + \sin{[\cos{(t)}]}$'

#     
ax04.plot(t, y, linewidth=1.5, linestyle='-', color='darkgreen', label=label_01)
ax04.plot(t, x, linewidth=1.0, linestyle='--', color='purple', label=label_02)

ax04.legend(ncol=2, title='', framealpha=0.7, fontsize=10)
plt.show()
```

```python
fig05 = plt.figure(num=4, figsize=(6, 8))  #   
axs05 = [fig05.add_subplot(3, 1, i) for i in range(1,4)]  #  

x = np.linspace(-2,2,100)
funcs = [x, x**2, x**3]
y_min = np.min(funcs)
y_max = np.max(funcs)

for (f, ax) in zip(funcs, axs05):
    ax.plot(x, f)
    ax.set_ylim(bottom=y_min, top=y_max)

#    Y     
for ax in axs05[1:]:
    ax.set_yticklabels([])

fig05.subplots_adjust(hspace=0)  #   hspace   
plt.show()
```

```python
fig08 = plt.figure(num=8)
ax08 = [
    fig08.add_axes([0.0, 0.0, 1.0, 1.0]),
    fig08.add_axes([0.20, 0.6, 0.2, 0.3], xlim=(-1.5, 0), ylim=(-3, 10)),
    fig08.add_axes([0.65, 0.6, 0.2, 0.3], xlim=(1.5, 2.1), ylim=(0, 10))
]

t = np.linspace(-2.0, 2.1, 500)
y = t**4 - t**3 + t**2 - t - 1
z = t**3 - t**2 + 3

for ax in ax08:
    ax.plot(t, y, linewidth=1.0)
    ax.plot(t, z, linewidth=1.0, markevery=10)
    #  set_xlabel  set_ylabel    

plt.show()
```

```python
fig09 = plt.figure(num=9)
ax09 = fig09.add_subplot(1, 1, 1)

t = np.linspace(-4*np.pi, 4*np.pi, 200)
y = np.sin(t)
ax09.plot(t, y)

#      
ax09.spines['top'].set_visible(False)
ax09.spines['left'].set_visible(False)

#   Ox   ,   Oy  
ax09.spines['bottom'].set_position('zero')
ax09.spines['right'].set_position('zero')

#       
ax09.spines['right'].set_visible(True)

#  
ax09.xaxis.set_tick_params(direction='inout', length=10)
ax09.yaxis.set_tick_params(direction='inout', length=10)

#   
ax09.yaxis.set_label_position("right")
ax09.yaxis.tick_right()

plt.show()
```
